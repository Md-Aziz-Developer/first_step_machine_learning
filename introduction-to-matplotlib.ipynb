## Matplotlib example workflow

```python
#import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x=[1,2,3,4]
y=[1,20,15,25]

#2. Setup Plot
fig,ax=plt.subplots(figsize=(10,10)) #width ,height

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple plot",xlabel="x-axis",ylabel="y-axis")

#5. Save and show (you save the whole image )
fig.savefig('images/sampleplot.png')
```

### Two options for subplots

```python
#Subplot option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2),
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(),nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplot option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should i use ? (pyplot vs matplotlib OO method)
* when ploting something quickly okey to use pyplot
* when ploting something more advance then OO method we use

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50
```

```python
over_50.head()
```

```python
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind='scatter',
            x='age',
             y='chol',
             c='target',
             figsize=(10,6)
            )
```

```python
#OO method mixed with pyolot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
# ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))
#plot the data
scatter=ax.scatter(x=over_50['age'],
                  y=over_50['chol'],
                  c=over_50['target'],
                   cmap='winter'
                  ) #cmap change the color scheme
# Customize the plot
ax.set(title="Heart Disease and Cholesterol level",
       xlabel='Age',
       ylabel='Cholesterol'
      );
#Add A lengend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--'
          );
```

```python
over_50.head()
```

```python
#subplot of age,chol,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10))
#Add data to ax0
scatter1=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                     c=over_50['target']
                   )
#Customize the plot ax0
ax0.set(title='Heart Disease and Cholesterol level',
       xlabel='Age',
        ylabel='Cholesterol'
       )
#Add Legend to ax0
ax0.legend(*scatter1.legend_elements(),title='Target')
#Add a horizontal line to ax0
ax0.axhline(over_50['chol'].mean(),linestyle='--');

#Add data to ax1
scatter2=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                     c=over_50['target']
                    )
#Customize the plot of ax1
ax1.set(title='Heart Disease and Max Heart Rate Level',
       xlabel='Age',
        ylabel='Heart Rate',
       )
#Add Legend to ax1
ax1.legend(*scatter2.legend_elements(),title='Target')
#Add horizonal line to ax1
ax1.axhline(over_50['thalach'].mean(),linestyle='--');

#Add mean heading to data
fig.suptitle('Heart Disease Analysis');
```

```python
#subplot of age,chol,thalach
#share x axis 
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10),sharex=True)
#Add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                     c=over_50['target']
                   )
#Customize the plot ax0
ax0.set(title='Heart Disease and Cholesterol level',
        ylabel='Cholesterol'
       )
#Add Legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target')
#Add a horizontal line to ax0
ax0.axhline(over_50['chol'].mean(),linestyle='--');

#Add data to ax1
scatter2=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                     c=over_50['target']
                    )
#Customize the plot of ax1
ax1.set(title='Heart Disease and Max Heart Rate Level',
       xlabel='Age',
        ylabel='Heart Rate',
       )
#Add Legend to ax1
ax1.legend(*scatter2.legend_elements(),title='Target')
#Add horizonal line to ax1
ax1.axhline(over_50['thalach'].mean(),linestyle='--');

#Add mean heading to data
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');
```
